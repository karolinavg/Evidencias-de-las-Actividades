# Anthropometry dataset

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
dat1 = '/content/drive/MyDrive/5 SEMESTRE/Visualization of Data for Decision Making/antropometria-dataset-2.csv'
```

```python
df=pd.read_csv(dat1)
df.head()
```

```python
type(dat1)
```

```python
df1 =df.copy()
df1
```

```python
df1.columns
```

```python
type(df1)
```

```python
pd.pivot_table(index=['desc_ent'],
                  aggfunc=[np.mean],
                  data = df1,
                  values=['peso'],
                   )
```

```python
df1.head()
```

```python
pd.pivot_table(index=['desc_ent'],
                  columns=['sexo'],
                  aggfunc=[np.median],
                  data = df1,
                  values=['peso'],
                   )
```

# Adult.csv

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
dat2='/content/drive/MyDrive/5 SEMESTRE/Visualization of Data for Decision Making/adult.csv'
```

```python
df2=pd.read_csv(dat2)
df2.head()
```

```python
pd.pivot_table(index=[df2['marital-status']=='Never-married',df2['education']=='Bachelors'],
                  aggfunc=[np.mean],
                  data = df2,
                  values='age',
                   )
```

```python
df2[df2.education == 'Bachelors'].pivot_table(index=['education'], 
                                                       values='age', 
                                                       aggfunc=[np.mean])
```
